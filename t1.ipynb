```python
#学习数据和绘制数据图
#做了一个简单的统计一个文本里各种词组数量的app
#后续会添加各种其他功能
#主要是为了以后研究伤寒和各种古代数据做准备

import re
import sys
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontManager
from pylab import mpl
import subprocess

plt.rcParams['figure.figsize'] = [18,13]#设置在网页上显示的大小

font2 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 20,
}


#这里填写需要统计的词，写在一个组里，任意多个
shanghan1 =['桂枝汤','四逆汤','汤','桂枝','甘草','姜','附子','发汗','恶寒','吐','死','伤寒','热','寒','下利','小青龙']
yuanjue1 = ['我','佛','幻','眾生','空','愛','無']




class Find_word():
    def __init__(self,w_list,w_file):
        self.file = w_file
        with open(self.file) as f:#读文件
            self.data = f.readlines()#一行一行的读
        self.list = w_list
        self.s_list = []
        

    def count(self,p):
        self.x = []
        self.a = []
        self.p = p
        self.n = 0

        for line in self.data:
                self.x = re.findall(self.p,line)#在美一行里查词
                if self.p in self.x:
                    self.a.append(self.x)#查到后装到list里

        for i in self.a:#从list里算到底有多少个这个词，但我觉得有更好但方法，请大神赐教
            if self.p in i:
                for j in i:
                     if self.p in j:
                        self.n += 1
        return self.n
        

    def show(self):
        plt.xlabel('目标词汇词')
        plt.title('数据分析')
        plt.xticks(fontsize=20)#x轴的字体大小

        for i in self.list:#这是把用户要查的词给放在图表的x轴上
            self.s_list.append(self.count(i))
            plt.text(i,self.count(i)+3,self.count(i),font2)

        plt.bar(self.list,self.s_list)#设置图表
        plt.show()#画出来



#填写需要统计的文档
pp = Find_word(shanghan1,'shanghan1.txt')
pp1 = Find_word(yuanjue1,'yuanjue.txt')


pp.show()
pp1.show()

```
