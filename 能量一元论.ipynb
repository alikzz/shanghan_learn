# 能量一元论

为了更好的表述张仲景"伤寒论"的核心理论, 因此做了一个叫做"能量一元论"的东西来简化和梳理张仲景对疾病,对人体,对治病的看法.
内部是为了更好的做程序而需要的一个指南.

所谓"能量一元论"并不是什么发明,只是将中国古人看待世界的奇特视角用现代人比较能接受的词语重新说一遍.

我对将任何复杂的事情简化有兴趣,对将简单的事情复杂化无兴趣.

一元论就是试图将中医系统内纷繁复杂,玄而又玄的东西给正常化,简单化,去神秘化.

好比当代大部分物理学家,都有一个梦想,就是找到能够对所有物理学领域的"大一统"理论.

因为人们相信, 如果这个世界是上帝所造, 那么一定有一个优雅而简单的理论,可以囊破世间的一切, 可以解释这看似复杂世间的一切.

对于中国的中医理论,我相信有一个理论可以将所有中医的门派,理论全部统一起来, 我不相信中医的核心思想是破碎的, 某一派只能只好某一种疾病, 某一本书和另外一本书不能兼容.

整个世界的一切是联系在一起, 中国古人从来都如此看, 具体到中医也不该例外, 而且尤其不该例外.
因为中医既是在探讨人体与整个世界的关系的学科.

中国向来有世间所有学问"一以贯之"的学术习惯, 我就试着用极简单的语言来描述这个"一".

## 只有能量
世间有且只有一个东西, 用当下流行的名词来说, 就是"能量".
甚至"时空"也是能量在转化中的副产品.

那么物质是什么?
物质是能量在不同时空的不同状态.
因为我们人类对时空认知的局限,我们会产生能量是能量,物质是物质的错觉.

能量既是物质,物质既是能量.

好比,一块木头, 和一团火.
他们是一个东西, 木头点燃后就会成为一团火.
而一团火, 在"历史"上曾经是一块木头.

## 一体多面

西方对能量既是能量在近代达成了共识, 爱因斯坦最有名的方程式, 就是他对物质和能量是一体的表述.

中国文化很早就发现这个事情,而且中国人做的更抽象.
我们这里不说那些玄乎的文字, 我尽量简单明快的说明.

我们给能量一个中国名字, 阳.
给物质一个中国名字, 阴.

但我们要永远牢记, 阴和阳不是分裂的, 他们只是在不同时空的不同表现而已, 他们是一个,不是两个.

好比, 白天和黑夜, 不是两天, 是一天.
白天会慢慢过渡成黑夜, 黑夜会慢慢过渡成白天.
因此没有一个"纯白天"和一个"纯黑夜", 尽管白天和黑夜看起截然相反.

两个是看起来对立, 实际是渐进互生互化的.
如果不明白这个, 那么下面的一切都没有意义.

## 能量与物质的互化
于是我们会看到能量在时空下在不断的做着转化, 一会能量物质化成了物质, 一会物质能量化成了能量.
这个过程从未停止, 如果停止了, 我们认为这是"死亡".

凡是有生命的个体, 无不是在做着能量互化, 中国古人称之为阴阳和合.

一块石头, 死了, 因为他长时间以物质的形态存在.
反观我们生物, 每分每秒都做着能量与物质的变化.

当我们不在阴阳互化的时候, 我们也就死了.

## 上升与下降

我们再把这个阴阳互化的过程放大一点看,看看这个过程具体是如何发生的.

我们还是想象一下木头生火这个过程.

当点燃火的时候, 火苗是永远向上的, 戴起来的灰尘也是向上的.

而热量给人的感觉就是向上,  相反, 把火熄灭, 一切凉下来, 就是一种向下的感觉.

我们看原子弹爆炸, 那个小原子弹变成巨大的蘑菇云, 就是物质变成能量的过程. 非常典型的向上. 当爆炸过后, 一切又慢慢冷却下降.

所以, 物质和能量内部互化只有两个趋势, 一个向上, 一个下降.

这两个趋势, 在世间一切物质上都有.

## 把上升和下降方的放大

刚才我们把能量转化的过程放大来看, 就看到了两个趋势, 一个向上, 一个向下.

我们会发现,这个确实可以自我闭合, 也就是形成一个闭合的圆, 然后这个过程就可以无限运转.

中国古人给这个闭合的无限运转的能与物互化的圆分成了五个特性,或是趋势.

中国古人管向上的势叫"木", 向下的叫"金", 上升达到顶点的叫"火",下降达到最低点的叫"水" ,中间那个像是车轴的, 叫"土".
这是中国古人一种形象的比喻, 为了拿这种存在于世间一切的普遍特性去概况和描述世间的一切.

"五行", 如果是现代人, 也许会这么称呼 "两极, 中轴, 向上, 向下", 古代人就换个名字,叫做"金木水火土".
所以五行是描述能量和物质互化过程的.

## 能量的世界

我们生活的这个世界, 充满了处在各个能量变化阶段的能量.
有的能量大, 有的能量小, 有的能量偏向寒冷, 有的能量偏向火热.

甚至我们的语言, 念头, 全部都是一种能量.
一首音乐, 一个电影, 都是一个不断变化, 有自己方向和能量等级的能量体.
这些东西不断和我们自己的能量做着交换和互动.

疾病其实本不是疾病, 他只是外部或内部的某种带有比较明显偏向的能量, 这个能量很强, 对我们身体的能量做出较大的改变.
当这个改变和我们身体想维持平衡的趋势僵持不下时, 就会停在某一处, 我们也许就能感受到疼痛, 或者舒服.

我们把那个给我们疼痛的结果, 叫做疾病.
所以并非一股寒冷的风会让我们生病, 一个恶意的语言, 或者某个想法都可能导致我们生病.

## 能量的两个属性

上面说了"五行",这是能量势头的细分, 其中那个"土"看起来有点奇怪, 他好像没有什么方向, 为什么古人把他放在五行里面,而且还放在中间呢?

好比一个打火机的能量和一个原子弹的能量, 这两者显然一个能量小, 一个能量大.
而"中土"是表示这个能量大小的. 至少在我要说的中医理论的五行中,是这个概念.

二, 是能量方向, 也就是"向上","向下","两极".

## 张仲景的六经辨证的六经到底说的是什么?
张老师说的六经, 用能量和能量属性来描述就特别简单了.

一, 能量的大小.
二, 能量的位置.

首先, 能量的大小指的是, 个患者自身能量的多寡, 张老师管自身能量多的(现代语言叫抵抗力)叫"阳",想法, 自身能量少的叫"阴".
所以就分阴阳两种病, 阳病指的就是患者能量还比较高, 那么当疾病进入能量身体高的人的治疗方法和能量低的患者治疗方法不一样,患者对疾病的反应也不一样, 一般来讲, 能量高的患者对疾病产生一种亢进的表现, 比如发热, 发汗,等等, 二能量低的患者的反应就是虚弱的, 比如嗜睡, 四肢冰冷 等等.

其次,确定了这个病人的阴阳属性后, 还要确定这个外来能量与字体能量在什么地方产生了抗衡.
这里张老师定义了三个阶段, 分别是"表,间,里".
这个很好理解, 就是病这个能量走到了身体的表层, 里层, 或者是在表层和里层之间.
因为这三个层次中患者也是有比较典型的反应, 以及针对这些反应的针对治疗.

所以, 这就是六经了.

自身能量比较多的病人, 疾病能量在表层, 在中间, 在内部.
自身能量比较少的病人, 疾病能量在表层, 在中间, 在内部.
是为六经.

按张老师的语言是:
太阳, 少阳, 阳明.
太阴, 厥阴, 少阴.

张老师就是用能量的多寡,以及能量相持的位置来确定"病"的"样子".

## 如何才有生命

现在我们接触到了"疾病"这个字眼.
前面说了, "疾病"就是外部某种比较强大的能量进入身体后打破身体平衡的反应, 身体是有一个非常强的自我保护机制的, 这其实也是某种特定的能量, 我将这种能量称之为"阳气".

阳气足的人, 第一较难生病, 因为外部的能量不容易进来.
其次就算进来, 因为阳气足, 自身能量多, 那么就有更多的力量将这个外部能量排出去.

那么当暂时无法排出去, 外部能量和自身能量僵持在那里了, 就会出现病痛等一系列反应, 我们就觉得我们病了.
不是发热, 就是发冷, 不是拉肚子就是吐, 不是头痛就是肌肉痛, 等等一系列症状.

我们后世的人, 管这个不良反应的结果叫"病".
那么中医治疗, 不放眼在如何把你的拉肚子给解决, 而是放眼在如何调整你自身的能量, 让你自身能量恢复, 重新建立起自我保护机制, 然后这个自我保护机制可以扫除一切所谓"病".

这个自我保护机制唯一做的, 就是恢复身体的阴阳平衡.

那什么是阴阳平衡呢?
我之前也老听说阴阳平衡, 但根本不理解到底是什么意思.

如果按照能量一元的看法看, 就很简单了.

上面说到, 死物就是不做阴阳变化的(能量物质变化)的, 那么活物的首要条件, 就是要可以做阴阳互化, 没有阴阳互化就不要谈生命.
那么既然生命有了, 什么才可以是健康的生命呢?

那么就是阴阳互化这个具体过程, 上升和下降是可以画成一个完美的圆.
就是这个过程可以一直不断持续下去, 什么东西可以持续, 只有圆, 而且越是完美的圆, 越是可以持续的运动.
一个方块滚两下就不动了, 而一个完美圆, 可以持续滚动下去.

这就是阴阳平衡.
其实非常简单, 举个例子.

## 为什么唯独地球有生命?

至今科学家也没能找到哪怕一个和地球一样的星球, 更没找到外星人.

原因很简单, 地球是这个宇宙里最阴阳平衡的星球.

如果地球偏向太阳方向哪怕万分之一, 那么地球上的水就会被蒸发掉.
如果地球偏向太阳相反方向, 那么地球的水就会被冰冻.

生命, 尤其是蓬勃的生命群, 必须生活在液态水的大环境下.
怎么才能有液态水?
就是不太冷, 也不太热, 恰好在一个特别适中的位置.

人体的健康也是一模一样.
人体身上大部分是液体, 不能过凉, 不能过热. 
两个极端都是病, 甚至都会死人.
人可以因寒而死, 可以因热而死.

而在中间的平衡地带, 就是健康, 就是持续的生命.

所以,健康的生命只有用阴阳平衡来形容是最贴切的.
只是因为这是古人的语境, 同时现在人们被西方的一切洗脑, 反而不能接受这种我认为最科学的解释了.

## 夸张的比喻

一个人站在火堆前伸出双手, 希望让手暖和起来.

但这个人不小心让手指离火堆太近了, 于是手指头被烤的非常疼.

这时来了两个医生, 一个是西医, 一个是中医.

西医这样给这个人的手痛治疗, 西医给手指头疼的地方打了麻药.
一个瞬间, 此人的手指头不疼了. 感觉病被治好了.但过一会又疼了, 所以需要不停的打麻药.

而中医这样治疗.
中医只是跟那个人说, 哎, 你的手离火太近了, 你把手拿开一点就好了.
没有打针也没有吃药, 同样, 这个人手痛被治好了.

我知道这个比喻有点夸张了, 但是实际上, 中西的思考方式就是这样.

中国古人考虑整体的阴阳平衡而已, 并且喜欢用最小的代价, 最好是什么都不做.
而西方有高超的科技, 对某一个问题可以用科技来解决, 但是他们却没有看清生命的整体.

再举一个比较实际的例子.

比如说高血压.西医认为是血管壁的问题, 或者是心脏的问题, 治疗的办法是终身服药, 这种化学药品可以让血压降下来.

而中医的思路不是这样.
首先, 中医理解, 需要高的源头是身体阴阳失衡, 更具体来讲是阴胜阳衰, 字体能量虚弱, 那么什么没那么多能量把营养物质和血压打到身体表层和大脑, 于是身体为了恢复这个平和, 做出了调整反应, 就是让身体血压升高, 弥补能量不足, 好将血液输送到身体表层.
那么治疗高血压, 跟治疗任何一个疾病都一样, 找到身体不平衡的地方, 用最小的代价使其平衡, 血压自然就好了.

我觉得中医治病其实最简单, 治疗感冒和治疗高血压其实一模一样, 本质没有任何区别.

因为中医抓住了主要矛盾, 能量, 以及能量的互化.

## 中医本质是能量管理学

如果我们一能量一元看世界, 那么一切活动都是能量转化.

无论是看手机, 听音乐, 玩游戏, 看电影, 吃饭, 睡觉, 做爱, 拉屎, 帮助, 憎恨, 等等全部是能量交换与能量互动.

因为中国古人看到这一切行为的内部运行法则, 并把这一切极度抽象和简化, 就产生了阴阳观和五行观.

最终都是能量一元观.

古人管这个能量叫气, 叫道, 等等.

中医因为看到了人类在这个大自然中受到的诸多影响, 以及明白自身能量运行规律, 于是可以用非常简单的方式调节能量平衡, 使之达到一个最优化组合.

中医就好比一个手机里的电池管理系统, 或者特斯拉电动汽车里的电池管理系统.

这个系统非常重要, 虽然不太起眼, 但是这甚至是整个产品的一个核心, 没有一个好的能源管理系统, 再好的软件, 再好的硬件也无法发挥.

我们来具体看看古人是如何治病的, 是如何运用整个管理系统的.

## 古人如何治病

![title](pic/yiyuan/15.png)

我们先回头看看古人眼中什么是病, 再谈他们如何治病.

首先, 我们自体能量变低, 无论外部能量还是以前内部的某些能量开始强烈的影响整体平衡.

在自体能量试图维持平衡时, 与外部能量僵持不下, 就产生了所谓的疾病.

这个疾病随着自体能量变低, 会逐步深入, 有表至间至里, 病情就是如此恶化.

说的在简单点, 病就是因为某种能量阻碍了身体阴阳平衡的循环, 因此产生的某种反应.

![title](pic/yiyuan/16.png)

相对应的, 中国古人的治疗方案是根据患者被阻碍的平衡, 以及偏离平衡的身体能量方向来入手.

医生只要将阻碍能量的排出体外, 以及将身体不阻碍身体能量平衡的错误方向逆转回来, 使身体重新回到能量的平衡转化循环中来, 那么自体有能力驱除体内一切所谓疾病.

自然界中有一股神奇的力量, 不只出现在人体身上, 也出现在一切事物上面.
这个力量就是万物都有趋于平衡的内在动力.

世界上的一切事物, 要想存在, 要想存在的长久, 尤其是生命, 好似必须处在平衡中.
平衡一旦被打破, 那么恢复平衡的机制立刻启动.

中医就是给恢复平衡助力而已.

所以理论上来讲, 中医不治病, 只恢复身体能量平衡, 治病的是身体本身.



## 用数据来揭开中医的神秘

正在做的伤寒论的全方位数据统计还没做出来, 但是我看到了一些大的数据情况, 很清晰的表明, 张仲景是如何恢复身体平衡的.

其实总结起来就两点:

1 增加体质 (伤寒中叫查津液之虚实, 中医中叫培育中土)
2 修改身体不正确的方向 (中医叫扶正)

看张仲景开的一百多个方子, 几乎都是这一个路数.

中土必须培育, 然后方向基本上是木的方向, 也就是向上的, 升阳的方向.

这就给我们一个古人心中健康的方向, 就是这个方向了.

我们人体每天就是不断的耗损, 睡眠, 情志, 环境, 冷暖在现代社会都非常的向冷向下,向过度消耗走.
那么无怪乎张仲景是如此用药了.
