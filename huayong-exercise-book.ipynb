```python
from shlib.shtext import ShangHanText
sht = ShangHanText()
```

```python
sht.load_text("data/SH_text.json")
```

```python
sht.search_keyword("恶风")
```

```python
import re
for item in re.finditer("发热", "太阳病。发热。汗出。恶风。脉缓者。名为中风。发热。热，发热发热"):
    print(item)
```

```python
word = "发热"
pattern = re.compile(word)
sht.text_dataframe.text.apply(lambda x: len(re.findall(pattern, x)))
```

```python
df = sht.text_dataframe[["six_channel", "text"]]
pattern = re.compile(word)
df["word_count"] = df.text.apply(lambda x: len(re.findall(pattern, x)))
count_in_six_channel = df.groupby("six_channel").sum()
count_in_six_channel
```
